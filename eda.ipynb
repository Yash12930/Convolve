```python
import pandas as pd
dev_data = pd.read_csv("Dev_data_to_be_shared.csv")
```

```python
bad_flag_1 = dev_data[dev_data['bad_flag'] == 1]
print(bad_flag_1.shape)
```

```python
mean_bad_flag_1 = bad_flag_1.mean()
mean_bad_flag_0 = dev_data[dev_data['bad_flag'] == 0].mean()

mean_difference = (mean_bad_flag_1 - mean_bad_flag_0).abs()
relevant_columns = mean_difference.sort_values(ascending=False)
print(relevant_columns.head(10))
```

```python
top_columns = relevant_columns.index[:10]
print("Top relevant columns:", top_columns)

import seaborn as sns
import matplotlib.pyplot as plt

for col in top_columns:
    plt.figure(figsize=(8, 4))
    sns.histplot(bad_flag_1[col], kde=True, bins=30)
    plt.title(f"Distribution of {col} for bad_flag == 1")
    plt.show()
```
